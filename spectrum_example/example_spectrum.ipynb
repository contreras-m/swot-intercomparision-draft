## User configuration

```python

start_date = "30042023" #"DDMMYYYY"
end_date ="10052023"
area = [11.5, -40.5, 20, -30.5] #[longitude_min, latitude_min, longitude_max, latitude_max]
sim_name="eORCA025"

# question: should define the cycle and pass number in function of time?
```

## Load data

```python
psd_dict, freqs_dict = sp.calculate_psd(datasets)
# Calculate PSD Mean
psd_mean, freqs_mean = sp.psd_mean_and_freq(psd_dict,freqs_dict)
```

## Figure

```python
fig, ax_ = plt.subplots(1, 1, figsize=(8, 8))
sp.plot_psd(ax_, freqs_mean, psd_mean, unit = "m2",psd_labels="PSD",title="PSD",savefig=False,plot_transparent=False)
```
